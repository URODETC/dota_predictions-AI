```python
from import_template import pred
import requests
```

```python
URL = "https://api.opendota.com/api"
league_id = 19099
req = requests.get(URL+f"/leagues/{league_id}/matches")
```

```python
response_per_minute = 50
id = [i.get("match_id",0) for i in req.json()][:response_per_minute]
```

```python
matches = [requests.get(URL+f"/matches/{i}").json() for i in id]
```

```python
sum([i["duration"]/60 for i in matches])/50
```

```python
sorted_players = [[[teams["hero_id"] for teams in [sorted(match["players"][i:i+5], key=lambda x: (x['lane_role'], x['gold_t'][10]))[[1,2,4,3,0][j]] for i in [0, 5] for j in range(5)]], match["radiant_win"]] for match in matches]
```

```python
sum([int(rw != round(pred.prediction(t[5:], t[:5])[5]["Radiant"])) for t, rw in sorted_players])/50
```
